1. Import library opencv, numpy, dan pyplot

```python
# Tulis kode Anda disini
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Upload sebuah citra, kemudian baca menggunakan opencv

```python
# Tulis kode Anda disini
img = cv2.imread("lambang umm.png")
```

3. Tampilkan ukuran dari citra tersebut

```python
# Tulis kode Anda disini
print(img.shape)
```

4. Ubah format dari citra tersebut dari BGR ke RGB

```python
# Tulis kode Anda disini
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

5. Tampilkan citra tersebut dengan menggunakan pyplot

```python
# Tulis kode Anda disini
plt.imshow(img_rgb)
```

6. Ubah ukuran citra tersebut menjadi 50% dari ukuran aslinya, tanpa mengubah aspect ratio nya. Kemudian tampilkan ukuran citra yang baru tersebut!

```python
# Tulis kode Anda disini
scale_percent = 50
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

img_resized = cv2.resize(img_rgb, dim, interpolation=cv2.INTER_AREA)

plt.imshow(img_resized)
print("Original shape: "  + str(img.shape))
print("Resized shape:" + str(img_resized.shape))
```

7. Lakukan rotate 75 derajat kepada citra asli. Kemudian tampilkan citra asli dan citra hasil rotate secara berdampingan!

```python
# Tulis kode Anda disini
import imutils
img_resized_rotated = imutils.rotate(img_resized, angle=75)

img_title = ['original', 'rotated']
imgs = [img_resized, img_resized_rotated]
count = 2
for i in range(count):
  plt.subplot(2, 2, i + 1)
  plt.title(img_title[i])
  plt.imshow(imgs[i])
```

8. Lakukan flip secara horizontal (kanan-kiri) kepada citra asli, kemudian tampilkan keduanya secara berdampingan!

```python
# Tulis kode Anda disini
img_resized_xflip = cv2.flip(img_resized, 1)

img_title = ['original', 'flipped']
imgs = [img_resized, img_resized_xflip]
count = 2
for i in range(count):
  plt.subplot(2, 2, i + 1)
  plt.title(img_title[i])
  plt.imshow(imgs[i])
```

9. Ubah format citra asli menjadi greyscale, kemudian tampilkan keduanya secara berdampingan!

```python
# Tulis kode Anda disini
img_resized_grayscaled = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

img_title = ['original', 'grayscaled']
imgs = [img_resized, img_resized_grayscaled]
count = 2
for i in range(count):
  plt.subplot(2, 2, i + 1)
  plt.title(img_title[i])
  plt.imshow(imgs[i], cmap='gray', vmin=0, vmax=255)
```

10. Simpan semua citra hasil perubahan Anda dalam file berformat JPG

```python
# Tulis kode Anda disini
cv2.imwrite("img_scaled.jpg", img_resized)
cv2.imwrite("img_rotated.jpg", img_resized_rotated)
cv2.imwrite("img_flipped.jpg", img_resized_xflip)
cv2.imwrite("img_grayscaled.jpg", img_resized_grayscaled)
```
